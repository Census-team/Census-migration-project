```python
import pandas  as pd
import numpy as np

moving_df = pd.read_excel("Moving_reasons.xlsx")

moving_df.head()
```

```python
moving_df.columns
```

```python
import altair as alt

alt.Chart(moving_df).mark_bar().encode(
    x="Mobility_Period",
    y="Total_per_year")
```

```python
moving_melt = pd.melt(moving_df, id_vars = "Mobility_Period",value_vars=['Change_in_marital_status',
       'To_establish_own_household', 'Other_family_reason',
       'New_job_or_job_transfer', 'To_look_for_work_or_lost_job',
       'To_be_closer_to_work/easier_commute', 'Retired',
       'Other_job-related_reason', 'Wanted_to_own_home,_not_rent',
       'Wanted_newer/better/larger_house_or_apartment',
       'Wanted_better_neighborhood/less_crime', 'Cheaper_housing',
       'Foreclosure_eviction', 'Other_housing_reason',
       'Relationship_with_unmarried_partner', 'To_attend_or_leave_college',
       'Change_of_climate', 'Health_reasons', 'Natural_disaster',
       'Other_reasons'], var_name = "Moving reason", value_name="Number who moved")

moving_melt.head()
```

```python
moving_melt["Number who moved"] = moving_melt["Number who moved"].replace('N', np.NaN)
```

```python
moving_melt["Number who moved"] = pd.to_numeric(moving_melt["Number who moved"])
```

```python
moving_melt.groupby(["Mobility_Period","Moving reason"]).agg({"Number who moved":np.mean})
```

```python
alt.Chart(moving_melt).mark_bar().encode(
x = alt.X("Moving reason", sort='-y'),
y = alt.Y( "Number who moved"),
column = "Mobility_Period"
)
#column = "Mobility_Period")
```
